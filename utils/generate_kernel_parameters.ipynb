```python
import os
from itertools import product
import pandas as pd
```

```python
eta = [i/2 for i in range(1, 21)]
power = [i/2 for i in range(1, 41)]

from itertools import product

kernel_types = ['exponential_kernel', 'lorentz_kernel', 'radial_basis_function', 'polynomial_kernel']
pairs = list(product(eta, power))  # 20 * 40 = 800 pairs

rows = [{'type': k, 'tau': e, 'power': p} for k in kernel_types for e, p in pairs]

import pandas as pd
df = pd.DataFrame(rows)
print(df.shape)  # should be (2400, 3)
df.to_csv('kernels.csv', index=False)
```
